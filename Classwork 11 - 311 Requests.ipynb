Create new columns called `created_at` and `closed_at` from the `Created Date` and `Closed Date` columns. Do them individually, and **time them**.

```python
%%time
df['created_at'] = pd.to_datetime(df['Created Date'].head(10000), format="%m/%d/%Y %H:%M:%S %p")
```

```python
%%time
df['closed_at'] =  pd.to_datetime(df['Closed Date'].head(10000), format="%m/%d/%Y %H:%M:%S %p")
```
